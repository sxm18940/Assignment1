```python
#1. Numpy:
#Using NumPy create random vector of size 15 having only Integers in the range 1-20.
#1.a Reshape the array to 3 by 5
import numpy as np
rand_vector = np.random.randint(1,20, size = 15).reshape(3,5)
print (rand_vector)
```

```python
#1.b Print array shape.
print("Array_Shape: ",rand_vector.shape)
```

```python
#1.c Replace the max in each row by 0
Max_val= np.amax(rand_vector, axis = 1)
Max_val= Max_val[:, None]
new_Arr = np.where(rand_vector==Max_val,0,rand_vector)
print(new_Arr)
```
